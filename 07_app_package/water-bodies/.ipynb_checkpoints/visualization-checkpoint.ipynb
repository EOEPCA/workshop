```python
def get_center_from_bbox(bbox):
    x1, y1, x2, y2 = bbox
    center_x = (x1 + x2) / 2
    center_y = (y1 + y2) / 2
    center = (center_x, center_y)
    return center
```

```python
bbox = it1.bbox
bbox
```

```python
bbox_proj = it1.properties['proj:bbox']
bbox_proj
```

```python
center = get_center_from_bbox(bbox)
print(center)
```

```python
import numpy as np
center2 = tuple(np.roll(center,1))
center2
```

```python
m = leafmap.Map(center=center2, zoom = 9)

for item in catalog.get_all_items():
    
    m.add_raster(item.get_assets()["data"].get_absolute_href(), layer_name=item.id)

m
```

```python
import rasterio
from rasterio.warp import calculate_default_transform, reproject, Resampling

path = it1.get_assets()['data'].href
# Open the raster file
with rasterio.open(path) as src:
    # Define the target CRS (EPSG:4326)
    dst_crs = 'EPSG:4326'

    # Calculate the transformation parameters
    transform, width, height = calculate_default_transform(src.crs, dst_crs, src.width, src.height, *src.bounds)

    # Create the output file and specify the transformation parameters
    kwargs = src.meta.copy()
    kwargs.update({
        'crs': dst_crs,
        'transform': transform,
        'width': width,
        'height': height
    })

    # Create the output file
    with rasterio.open("test.tif", 'w', **kwargs) as dst:
        # Perform the reprojection
        reproject(
            source=rasterio.band(src, 1),
            destination=rasterio.band(dst, 1),
            src_transform=src.transform,
            src_crs=src.crs,
            dst_transform=transform,
            dst_crs=dst_crs,
            resampling=Resampling.nearest
        )
```
